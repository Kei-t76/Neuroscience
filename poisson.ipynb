```python
#ポワソン過程発火モデル
#発火率4Hz, 発火回数30回の時系列データを計算

import numpy as np
import random
```

```python
dt = 1e-5   #時間幅
spike = 30  #30回のスパイクを生成
time = []   #スパイク時系列
n = 0   #スパイクが出た回数
t = 0   #時間
```

```python
while (n < spike):
    if random.random() < 4e-5: #スパイク発火
        n = n + 1
        time.append(dt*t)
    t = t + 1
```
